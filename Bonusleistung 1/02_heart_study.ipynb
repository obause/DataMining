## 2.1 Datenbereinigung

Folgende Bereinigungsschritte wurden durchgefÃ¼hrt:    

1. Geschlecht vereinheitlichen
2. Abiturbezeichnung angleichen
3. Datensatz 19 lÃ¶schen
4. Datensatz 6 lÃ¶schen
5. BPMeds Spalte lÃ¶schen
6. hadStroke Spalte lÃ¶schen
7. diabetes Spalte lÃ¶schen

```python
df = df_orig.copy()
# 1. Geschlecht vereinheitlichen
# Da nicht zu 100% davon ausgegangen werden kann, dass die Merhzahl an Rauchern MÃ¤nner sind wurde der Datensatz gelÃ¶scht
# df.replace({'f':'1'}, inplace=True)
df.drop(11, inplace=True)

# 2. Abiturbezeichnung angleichen
df.loc[df['educational background'] == 'Abituriat', ['educational background']] = 'Abitur'

# 3. Datensatz 19 lÃ¶schen
df.drop(19, inplace=True)

# 4. Datensatz 6 lÃ¶schen
df.drop(6, inplace=True)

# 5. BPMeds Spalte lÃ¶schen
# 6. hadStroke Spalte lÃ¶schen
# 7. diabetes Spalte lÃ¶schen
df.drop(columns=['BPMeds', 'hadStroke', 'diabetes'], inplace=True)
```

## 2.2 Datentransformation 1: Diskretisierung
Numerische Werte in Wertebereiche umwandeln
Neue AusprÃ¤gungen: 

- Alter in (jung, alt)
- Zigaretten am Tag in (kein, wenig, viel)
- BMI in (Untergewicht, Normalgewicht, Ãœbergewicht)


```python
dfBins = df.copy()
bin_labels_alt = ['jung', 'alt']
dfBins['age'] = pd.cut(x = df['age'],
                              bins=[0, 40, 100],
                              labels = bin_labels_alt)

bin_labels_rauchen = ['kein', 'wenig', 'viel']
dfBins['cigsPerDay'] = pd.cut(x = df['cigsPerDay'],
                              bins=[-1, 1, 15, 100],
                              labels = bin_labels_rauchen)

bin_labels_bmi = ['Untergewicht', 'Normalgewicht', 'Ãœbergewicht', 'starkes Ãœbergewicht']
dfBins['BMI'] = pd.cut(x = df['BMI'],
                              bins=[0, 18.5, 25, 35, 100],
                              labels = bin_labels_bmi)#
dfBins
```

## 2.3 Datentransformation 2: Numerische Kodierung und Normierung
Nun sollen alle Daten in normierte numerische Werte transformiert werden. Legen Sie auch hierfÃ¼r eine Kopie "dfNumeric" des originÃ¤ren bereinigten Dataframes an.
1. AusprÃ¤gungen von Geschlecht, Position und Bildung durch numerische aufeinanderfolgende Werte (1,2,...) ersetzen. 
2. Werte auf das Intervall [0, 1] normalisieren. HierfÃ¼r wird der MinMaxScaler() verwendet, den Sie oben importiert haben.

Lassen Sie sich erneut das Ergebnis ausgeben.

```python
dfNumeric  = df.copy()
dfNumeric['educational background'].replace(to_replace=['Ausbildung', 'Abitur', 'Bachelor', 'Master'], value=[1, 2, 3, 4], inplace=True)
dfNumeric['Smoker'].replace(to_replace=['no', 'yes'], value=[0, 1], inplace=True)
dfNumeric.head()
```

```python
scaler = MinMaxScaler()

dfNumeric = pd.DataFrame(scaler.fit_transform(dfNumeric), columns = dfNumeric.columns, index = dfNumeric.index)
```
